```python
# Convert to NumPy arrays
train_x = np.array(train_x)
train_y = np.array(train_y)


val_x = np.array(val_x)
val_y = np.array(val_y)
# you have to make label categorical cross entropy

from tensorflow.keras.utils import to_categorical

# One-hot encode the labels
train_y_one_hot = to_categorical(train_y-1)
val_y_one_hot = to_categorical(val_y-1)
```

```python


# # Features (2D coordinates)
# X = [[2.1, 3.4], [1.8, 3.2], [5.4, 7.3], [5.7, 6.9], [8.5, 9.1], [8.3, 8.9]]

# # Labels (Classes 1, 2, 3)
# y = [1, 1, 2, 2, 3, 3]

# # Subtract 1 to make labels 0-based (optional, only needed for sparse categorical cross-entropy)
# y = [label - 1 for label in y]

# Define the model
model = Sequential([
    Dense(2, activation='relu', input_shape=(2,)),
    Dense(6 , activation='relu'),
    Dense(3, activation='softmax')  # 3 classes
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',  # Integer labels
    metrics=['accuracy']
)

# Train the model
result_of_model = model.fit(train_x, train_y_one_hot, epochs=50, batch_size=10, verbose=1, validation_data = (val_x , val_y_one_hot) )
```

```python

# Extract loss and accuracy from history
train_loss = result_of_model.history['loss']
val_loss = result_of_model.history['val_loss']
train_acc = result_of_model.history.get('accuracy')  # For training accuracy (may be 'acc' in older versions)
val_acc = result_of_model.history.get('val_accuracy')  # For validation accuracy


epochs = range(1, len(train_loss) + 1)
```

```python
# Plot Training and Validation Loss
plt.figure(figsize=(12, 6))

plt.plot(epochs, train_loss, label='Training Loss', marker='o')
plt.plot(epochs, val_loss, label='Validation Loss', marker='o')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
```

```python
test_y = np.array(test_y)
test_x = np.array(test_x)
test_y_one_hot = to_categorical(test_y-1)
```

```python
test_loss, test_accuracy = model.evaluate(test_x, test_y_one_hot, verbose=1)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
predictions = model.predict(test_x)
predicted_classes = predictions.argmax(axis=1)  # Get class with highest probability
print("Predicted classes:", predicted_classes)
print("Actual classes:", test_y-1)
```

Creating decision region plot

```python
# Generate a mesh grid over the feature space
x_min, x_max = train_x[:, 0].min() - 1, train_x[:, 0].max() + 1
y_min, y_max = train_x[:, 1].min() - 1, train_x[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

# Predict class for each point in the grid
grid_points = np.c_[xx.ravel(), yy.ravel()]
Z = model.predict(grid_points)
Z = np.argmax(Z, axis=1)  # Convert probabilities to class labels
Z = Z.reshape(xx.shape)

# Plot decision regions
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdYlBu)

# Overlay training data
scatter = plt.scatter(train_x[:, 0], train_x[:, 1], c=train_y, edgecolor='k', cmap=plt.cm.RdYlBu)
plt.colorbar(scatter)

# Plot details
plt.title("Decision Regions and Training Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
